```python
#Q10


# Load the dataset
df = pd.read_csv('BIKE DETAILS.csv')  # Replace 'dataset.csv' with your actual file path

# Filter data for first-owner bikes
first_owner_bikes = df[df['owner'] == 'First Owner']

# Create a scatter plot to show the relationship between selling_price and km_driven
sns.scatterplot(data=first_owner_bikes, x='km_driven', y='selling_price')
plt.title('Relationship between Selling Price and KM Driven for First Owner Bikes')
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price')
plt.show()
```

```python
#Q11

# Load the dataset
df = pd.read_csv('BIKE DETAILS.csv')  # Replace 'dataset.csv' with your actual file path

# Calculate the IQR for km_driven column
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

# Identify and remove outliers in the km_driven column
filtered_df = df[~((df['km_driven'] < (Q1 - 1.5 * IQR)) | (df['km_driven'] > (Q3 + 1.5 * IQR)))]

# Show the filtered dataset without outliers
print(f"Data after removing outliers in km_driven:\n{filtered_df}")
```

```python
#Q13

# Load the dataset
df = pd.read_csv('BIKE DETAILS.csv')  # Replace 'dataset.csv' with your actual file path

# Calculate the bike's age (current year - manufacturing year)
current_year = 2024  # Replace with the current year if necessary
df['age'] = current_year - df['year']

# Calculate the average depreciation in selling price based on the bike's age
average_depreciation = df.groupby('age')['selling_price'].mean()

print("The average depreciation in selling price based on the bike's age:")
print(average_depreciation)
```

```python
#Q14

# Load the dataset
df = pd.read_csv('BIKE DETAILS.csv')  # Replace 'dataset.csv' with your actual file path

# Calculate the average selling price for each manufacturing year
average_price_per_year = df.groupby('year')['selling_price'].mean()

# Find bikes priced significantly above the average price for their year
df['above_avg'] = df.apply(lambda row: row['selling_price'] > average_price_per_year[row['year']], axis=1)

# Get the bike names that are priced significantly above average
bikes_above_avg = df[df['above_avg']]['name'].unique()

print("The bike names priced significantly above the average price for their manufacturing year are:")
print(bikes_above_avg)
```

```python


# Load the dataset
df = pd.read_csv('BIKE DETAILS.csv')  # Replace 'dataset.csv' with your actual file path

# Select only the numeric columns for the correlation matrix
numeric_columns = df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix for numeric columns
correlation_matrix = df[numeric_columns].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))  # Adjust the size of the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numeric Columns')
plt.show()
```
