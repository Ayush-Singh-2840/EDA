```python
#Q1

import pandas as pd

# CSV file load karna
df = pd.read_csv('Car Sale.csv')

# Har dealer ke liye average selling price calculate karna
average_price_per_dealer = df.groupby('Dealer_Name')['Price ($)'].mean()

# Average prices ko descending order mein sort karna
average_price_per_dealer_sorted = average_price_per_dealer.sort_values(ascending=False)

# Results print karna
print("Average Selling Price per Dealer:")
print(average_price_per_dealer_sorted)

# Plot karne ka option agar aap compare karna chahte hain
import matplotlib.pyplot as plt

average_price_per_dealer_sorted.plot(kind='bar', figsize=(12, 6))
plt.title('Average Selling Price per Dealer')
plt.xlabel('Dealer Name')
plt.ylabel('Average Selling Price ($)')
plt.show()
```

```python
#Q2
df = pd.read_csv('Car Sale.csv')

# Group the data by Company and calculate the standard deviation of prices
price_variation_per_company = df.groupby('Company')['Price ($)'].std()

# Find the company with the highest price variation
max_variation_company = price_variation_per_company.idxmax()
max_variation_value = price_variation_per_company.max()

print(f"The company with the highest price variation is {max_variation_company} with a standard deviation of {max_variation_value:.2f}.")

# Plotting the price variation for each company
price_variation_per_company_sorted = price_variation_per_company.sort_values(ascending=False)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
price_variation_per_company_sorted.plot(kind='bar', color='orange')
plt.title('Price Variation (Standard Deviation) per Car Brand')
plt.xlabel('Company')
plt.ylabel('Standard Deviation of Price ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Q3

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Car Sale.csv')

# Function to calculate IQR
def calculate_iqr(prices):
    Q1 = prices.quantile(0.25)
    Q3 = prices.quantile(0.75)
    IQR = Q3 - Q1
    return IQR

# Group data by Transmission and calculate IQR for Price
iqr_per_transmission = df.groupby('Transmission')['Price ($)'].apply(calculate_iqr)

# Print IQRs for each transmission type
print("Interquartile Range (IQR) for Car Prices by Transmission Type:")
print(iqr_per_transmission)

# Boxplot to visualize the distribution of car prices for each transmission type
plt.figure(figsize=(10, 6))
df.boxplot(column='Price ($)', by='Transmission', grid=False)
plt.title('Distribution of Car Prices by Transmission Type')
plt.suptitle('')  # Remove the default boxplot title
plt.xlabel('Transmission Type')
plt.ylabel('Price ($)')
plt.show()
```

```python
#Q12

# Load the CSV file
df = pd.read_csv('Car Sale.csv')

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract the month and year from the date
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Count the number of cars sold per month (or per year if needed)
sales_by_month = df.groupby('Month')['Car_id'].count()

# Optionally, group by Year and Month to analyze annual trends
sales_by_month_year = df.groupby(['Year', 'Month'])['Car_id'].count().unstack()

# Plot the monthly sales trend
plt.figure(figsize=(10, 6))
sales_by_month.plot(kind='bar', color='lightblue')
plt.title('Car Sales Trend by Month')
plt.xlabel('Month')
plt.ylabel('Number of Cars Sold')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Plot the sales trend by Year and Month to see the seasonal trend across multiple years
plt.figure(figsize=(12, 8))
sales_by_month_year.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Car Sales Trend by Year and Month')
plt.xlabel('Year and Month')
plt.ylabel('Number of Cars Sold')
plt.tight_layout()
plt.show()
```

```python
#Q14
import seaborn as sns

# Load the CSV file
df = pd.read_csv('Car Sale.csv')

# Clean the 'Engine' column by mapping text to numeric values
engine_mapping = {'V6': 6, 'I4': 4, 'V8': 8, 'I6': 6, 'I5': 5}
df['Engine'] = df['Engine'].map(engine_mapping)

# Check for missing or invalid 'Engine' values and handle them
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')  # Convert to NaN for invalid values

# If there are NaN values in the 'Engine' column, check and fill with the mode or a fallback value
if df['Engine'].isna().any():
    mode_engine = df['Engine'].mode()
    if not mode_engine.empty:
        df['Engine'] = df['Engine'].fillna(mode_engine[0])  # Fill NaN with the most frequent value
    else:
        df['Engine'] = df['Engine'].fillna(df['Engine'].median())  # If no mode, fill with median

# Clean the 'Annual Income' column, ensuring it's numeric
df['Annual Income'] = pd.to_numeric(df['Annual Income'], errors='coerce')  # Convert to NaN for invalid values
df['Annual Income'] = df['Annual Income'].fillna(df['Annual Income'].mean())  # Fill NaN with the mean

# Calculate correlations between car price, engine size, and annual income
correlation_matrix = df[['Price ($)', 'Engine', 'Annual Income']].corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Between Car Price, Engine Size, and Annual Income')
plt.tight_layout()
plt.show()

# Scatter plot to visualize the interaction between these variables
plt.figure(figsize=(14, 8))

# Scatter plot between Price and Engine size
plt.subplot(1, 2, 1)
plt.scatter(df['Engine'], df['Price ($)'], alpha=0.5, color='blue')
plt.title('Car Price vs. Engine Size')
plt.xlabel('Engine Size')
plt.ylabel('Price ($)')

# Scatter plot between Price and Annual Income
plt.subplot(1, 2, 2)
plt.scatter(df['Annual Income'], df['Price ($)'], alpha=0.5, color='green')
plt.title('Car Price vs. Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Price ($)')

plt.tight_layout()
plt.show()
```

```python
#Q15

# Load the CSV file
df = pd.read_csv('Car Sale.csv')

# Check the unique values in the 'Engine' column before cleaning
print("Engine Column Before Cleaning:")
print(df['Engine'].unique())

# Update the engine column mapping to handle the new engine types
engine_mapping = {
    'Double\xa0Overhead Camshaft': 'DOHC',
    'Overhead Camshaft': 'OHC'
}
df['Engine'] = df['Engine'].map(engine_mapping)

# Check if the mapping worked as expected
print("Engine Column After Mapping:")
print(df['Engine'].unique())

# Fill NaN values in 'Engine' column with a placeholder or the most common engine type
df['Engine'] = df['Engine'].fillna('Unknown')

# Calculate the average price for each combination of car model and engine type
average_price_by_model_engine = df.groupby(['Model', 'Engine'])['Price ($)'].mean().reset_index()

# Check the result of the grouping operation
print("Average Price by Model and Engine Type:")
print(average_price_by_model_engine)

# Visualize the results with a heatmap for better insight
plt.figure(figsize=(12, 8))

# Ensure that the pivot operation works
pivot_table = average_price_by_model_engine.pivot_table(index='Model', columns='Engine', values='Price ($)', aggfunc='mean')

# Plot the heatmap
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Average Car Price by Model and Engine Type')
plt.xlabel('Engine Type')
plt.ylabel('Car Model')
plt.tight_layout()
plt.show()
```
