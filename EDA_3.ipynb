```python
#Q1

import pandas as pd

# Load the CSV file
df = pd.read_csv('amazon.csv')

# Convert 'rating' column to numeric, forcing errors to NaN (if any)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate the average rating for each product category
average_rating_by_category = df.groupby('category')['rating'].mean().reset_index()

# Display the results
print(average_rating_by_category)
```

```python
#Q2

# Load the CSV file
df = pd.read_csv('amazon.csv')

# Convert 'rating_count' to numeric, forcing non-numeric entries to NaN
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Drop rows with NaN values in 'rating_count' column
df_cleaned = df.dropna(subset=['rating_count'])

# Group by category and get the top product by rating_count within each category
top_rating_count_by_category = df_cleaned.loc[df_cleaned.groupby('category')['rating_count'].idxmax()]

# Display the results
print(top_rating_count_by_category[['category', 'product_name', 'rating_count']])
```

```python
#Q3
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('amazon.csv')

# Check the first few rows to ensure the columns are correctly read
print("First few rows of the dataset:")
print(df.head())

# Check the column names to confirm proper loading
print("\nColumn names:")
print(df.columns)

# Check data types to ensure proper conversion
print("\nData types of the columns:")
print(df.dtypes)

# Convert columns to numeric (ignore errors to prevent crashes)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Check for any NaN values in both columns
print("\nChecking for NaN values in discounted_price or actual_price columns:")
print(df[['discounted_price', 'actual_price']].isna().sum())

# Drop rows with NaN values for these columns
df_clean = df.dropna(subset=['discounted_price', 'actual_price'])

# Check the first few rows of cleaned data
print("\nCleaned Data (first few rows):")
print(df_clean[['discounted_price', 'actual_price']].head())

# Plotting the distributions
plt.figure(figsize=(12, 6))

# Try a simpler plot: Histogram for both columns to inspect distributions
sns.boxplot(df_clean['discounted_price'], bins=50, alpha=0.5, label='Discounted Price', color='blue')
sns.boxplot(df_clean['actual_price'], bins=50, alpha=0.5, label='Actual Price', color='red')

# Titles and labels
plt.title('Distribution of Discounted Prices vs Actual Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#Q4

# Load the dataset
df = pd.read_csv('amazon.csv')

# Ensure that discount_percentage is numeric (handle errors as NaN)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Check if there are any NaN values in the discount_percentage column
print(f"Checking for missing values in 'discount_percentage': {df['discount_percentage'].isna().sum()}")

# Group by category and calculate the average discount percentage
average_discount_by_category = df.groupby('category')['discount_percentage'].mean().reset_index()

# Display the results
print("Average discount percentage by category:")
print(average_discount_by_category)

# Optionally, you can visualize the results as a bar plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='category', y='discount_percentage', data=average_discount_by_category, palette='viridis')
plt.title('Average Discount Percentage by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Discount Percentage (%)')
plt.xticks(rotation=90)  # Rotate x-axis labels if necessary
plt.show()
```

```python
#Q5

# Load the dataset
df = pd.read_csv('amazon.csv')

# Check for missing values in the 'rating_count' column
print(f"Checking for missing values in 'rating_count': {df['rating_count'].isna().sum()}")

# Make sure that 'rating_count' is numeric
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Group by product_name and get the product with the highest rating count
top_products = df.groupby('product_name')['rating_count'].max().reset_index()

# Sort the products by rating_count in descending order to get the most popular ones
top_products_sorted = top_products.sort_values(by='rating_count', ascending=False)

# Display the top 10 most popular product names
print("Top 10 most popular product names based on rating count:")
print(top_products_sorted.head(10))

# Optionally, you can visualize the top 10 products in a bar plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='rating_count', y='product_name', data=top_products_sorted.head(10), palette='viridis')
plt.title('Top 10 Most Popular Products by Rating Count')
plt.xlabel('Rating Count')
plt.ylabel('Product Name')
plt.show()
```

```python
#Q6
import pandas as pd
from collections import Counter
import re
import nltk
nltk.download('stopwords')
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.corpus import stopwords

# Load the dataset
df = pd.read_csv('amazon.csv')

# Combine product_name and about_product into one text column for keyword extraction
df['text'] = df['product_name'].fillna('') + ' ' + df['about_product'].fillna('')

# Clean the text: remove non-alphanumeric characters, convert to lowercase
df['text_clean'] = df['text'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x.lower()))

# Split the text into words
words = ' '.join(df['text_clean']).split()

# Remove stopwords (commonly used words like 'the', 'and', etc. that are not meaningful for analysis)
stop_words = set(stopwords.words('english'))
words_clean = [word for word in words if word not in stop_words]

# Count the frequency of each word
word_counts = Counter(words_clean)

# Convert the counter into a DataFrame
word_df = pd.DataFrame(word_counts.items(), columns=['Keyword', 'Frequency'])

# Sort by frequency in descending order
word_df_sorted = word_df.sort_values(by='Frequency', ascending=False)

# Display the top 10 most popular keywords
print("Top 10 most popular product keywords:")
print(word_df_sorted.head(10))

# Visualize the top 10 keywords in a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Frequency', y='Keyword', data=word_df_sorted.head(10), palette='viridis')
plt.title('Top 10 Most Popular Product Keywords')
plt.xlabel('Frequency')
plt.ylabel('Keyword')
plt.show()
```

```python
#Q7
from collections import Counter
import re
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('amazon.csv')

# Combine review_title and review_content into one text column for review keyword extraction
df['review_text'] = df['review_title'].fillna('') + ' ' + df['review_content'].fillna('')

# Clean the review text: remove non-alphanumeric characters, convert to lowercase
df['review_text_clean'] = df['review_text'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x.lower()))

# Split the text into words
words = ' '.join(df['review_text_clean']).split()

# Count the frequency of each word
word_counts = Counter(words)

# Convert the counter into a DataFrame
word_df = pd.DataFrame(word_counts.items(), columns=['Word', 'Frequency'])

# Sort by frequency in descending order
word_df_sorted = word_df.sort_values(by='Frequency', ascending=False)

# Display the top 10 most popular words in reviews
print("Top 10 most popular product review words:")
print(word_df_sorted.head(10))

# Visualize the top 10 words in reviews in a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Frequency', y='Word', data=word_df_sorted.head(10), palette='viridis')
plt.title('Top 10 Most Popular Words in Product Reviews')
plt.xlabel('Frequency')
plt.ylabel('Word')
plt.show()
```

```python
#Q8

# Load the dataset
df = pd.read_csv('amazon.csv')

# Clean the 'discounted_price' column by removing currency symbols and converting to numeric
df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)

# Clean the 'rating' column by converting it to numeric (handle any non-numeric values)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate the correlation between 'discounted_price' and 'rating'
correlation = df[['discounted_price', 'rating']].corr()

# Display the correlation matrix
print("Correlation between discounted_price and rating:")
print(correlation)
```

```python
#Q9

# Load the dataset
df = pd.read_csv('amazon.csv')

# Clean the 'rating' column by converting it to numeric (handle any non-numeric values)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Group by category and calculate the average rating for each category
avg_rating_by_category = df.groupby('category')['rating'].mean()

# Sort the categories by average rating in descending order and select the top 5
top_5_categories = avg_rating_by_category.sort_values(ascending=False).head(5)

# Display the results
print("Top 5 categories based on highest average ratings:")
print(top_5_categories)
```

```python
#Q10
# Load the dataset
df = pd.read_csv('amazon.csv')

# Clean 'discounted_price' column
df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': '', '|': ''}, regex=True)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Clean 'rating' column (assuming ratings are in numeric format, if not convert similarly)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate correlation between 'discounted_price' and 'rating'
correlation = df[['discounted_price', 'rating']].corr()

# Display correlation matrix
print("Correlation between discounted_price and rating:")
print(correlation)

# Plot the correlation
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between Discounted Price and Rating')
plt.show()
```
